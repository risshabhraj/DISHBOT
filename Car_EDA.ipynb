**2-> Dropping Irrelevant columns**

```
car_data = car_data.drop(['Market Category', 'Engine Fuel Type'], axis=1)
car_data.head()
```

**3-> Renaming the Columns**

```
car_data = car_data.rename(columns = {"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
car_data
```

 **4-> Checking for any duplicates in the data, checking for null values and missing data.**

```
car_data.isnull()
```

```
car_data.isnull().sum()
```

**4.1-> Filling the mising data.**

```
car_data['HP'].fillna(car_data['HP'].median(), inplace = True)
```

```
car_data['Cylinders'].fillna(car_data['Cylinders'].median(), inplace = True)
```

```
car_data['Number of Doors'].fillna(car_data['Number of Doors'].median(), inplace = True)
```

```
car_data.isnull().sum()
```

**Finding the max MSRP each company has for its models**

```
price_data = car_data.groupby('Make')['Price'].max().reset_index()
price_data
```

```
car_data1['price_group'] = pd.cut(car_data1['Price'],[0,20000,40000,60000,80000,100000,600000], labels=['<20k','20-39k','40-59k','60-79k','80-99k','>100k'], include_lowest= True)
car_data1['price_group'] = car_data1['price_group'].astype(object)
```

```
(car_data1['price_group'].value_counts()/len(car_data1)*100).plot.bar(figsize=(10,6))
plt.title("Price group bar diagram")
plt.ylabel('% of Vehicles')
plt.xlabel('Price Group')
```
